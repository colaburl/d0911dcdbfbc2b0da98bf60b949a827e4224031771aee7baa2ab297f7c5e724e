```python
retr = pt.TerrierRetrieve.from_dataset('vaswani', 'terrier_stemmed', wmodel='BM25', num_results=5)
```

```python
pipeline = retr >> pt.text.get_text(pt.get_dataset('irds:vaswani'), 'text')
```
